```python
def unnest(df, col):
    unnested = (df.apply(lambda x: pd.Series(x[col]), axis=1)
                .stack()
                .reset_index(level=1, drop=True))
    unnested.name = col
    return df.drop(col, axis=1).join(unnested)
```

```python
OUT =  OUT.reset_index(level = 1 )
OUT = OUT.reset_index(level = 1)
```
